```python
# Gas Analysis part 1
df5 = pd.read_csv('GasAnalysis1Output.txt', sep = '       ', header=None, engine='python') # doesn't show first line; So use header = None it'll count the first line too!!
df5.columns = ['yearMonth', 'averageGasPrice']
df5['averageGasPrice'] = df5['averageGasPrice']/1000000000
df5['yearMonth'] = df5['yearMonth'].str.replace('"', '')
df5[['Month','Year']] = df5.yearMonth.str.split("/",expand=True)
#display(df5)
df5 = df5.sort_values(['Year', 'Month'], ascending = (True, True))

df5.groupby(['Year']).sum().plot(rot=70)
df5.groupby(['Year', 'Month']).sum().plot(rot=70)
```

```python
# fork analysis #2017 -10 # The Byzantium fork
df6 = pd.read_csv('fork1Output.txt', sep = '      ', header=None, engine='python') # doesn't show first line; So use header = None it'll count the first line too!!
df6.columns = ['Day', 'Value']
df6['Value']=df6['Value'].str.replace("\[", "")
df6['Value']=df6['Value'].str.replace("\]", "")
df6[['Total_transactions','Total_gasPrice']] = df6.Value.str.split(",",expand=True)
df6.drop('Value', axis=1, inplace=True)
df6 = df6.sort_values('Day')
df6['Total_transactions'] = pd.to_numeric(df6['Total_transactions'])
df6['Total_gasPrice'] = pd.to_numeric(df6['Total_gasPrice'])/100000
display(df6)
```

```python
df6.plot(x ='Day', y='Total_gasPrice', kind = 'line')
plt.show()

df6.plot(x ='Day', y='Total_transactions', kind = 'line')
plt.show()

df6['AveragePrice'] = df6['Total_gasPrice']/df6['Total_transactions']
df6.plot(x ='Day', y='AveragePrice', kind = 'line')
plt.show()
```

```python
# Gas Analysis 2 
df7 = pd.read_csv('GasAnalysis2Output.txt', sep = '     ', header=None, engine='python')
df7.columns = ['Date', 'Values']
df7['Date'] = df7['Date'].str.replace("\[\"", "")
df7['Date'] = df7['Date'].str.replace("\"\]", "")
df7['Values'] = df7['Values'].str.replace("\[", "")
df7['Values'] = df7['Values'].str.replace("\]", "")
df7[['Year','Month']] = df7.Date.str.split("-",expand=True)
df7[['Difficulty','GasUsed']] = df7.Values.str.split(",",expand=True)
# df7 = df2.sort_values(['Year', 'Month'], ascending = (True, True))
df7['Difficulty']=pd.to_numeric(df7['Difficulty'])
df7['Difficulty']=df7['Difficulty']/1000000
df7['GasUsed']=pd.to_numeric(df7['GasUsed'])
df7 = df7.sort_values(['Year', 'Month'], ascending = (True, True))
df7[0:]
```

```python
df7.plot(x ='Year', y='Difficulty', kind = 'line')
plt.show()

df7.plot(x ='Year', y='GasUsed', kind = 'line')
plt.show()
```
